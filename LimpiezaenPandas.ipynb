```python
import pandas as pd
```

```python
df = pd.read_csv("lago_aulas_4_1813327__20230904-5548-1d0aawt.csv", error_bad_lines=False, nrows=None)
```

```python
# Mostrar las dimensiones del DataFrame
print("Dimensiones del DataFrame:", df.shape)
```

```python
# Nombres de las columnas
print("Nombres de las columnas:", df.columns.tolist())
```

```python
# Tipos de datos de las columnas
print("Tipos de datos de las columnas:\n", df.dtypes)
```

```python
df.head(10)
```

```python
# elimine las columnas de longitud, latitud y la de humedad que (segundo la persona que toma los datos me explico que este dato no es tan confiable por que el sensor esta dentro de una cajita)
# estas columnas no me aportan valor dentro del analisis posterior de los datos obtenidos
# Lista de nombres de columnas a eliminar
columnas_a_eliminar = ['Unnamed: 4', 'Unnamed: 3', 'Sensor_Package_Name.4']

# Eliminar las columnas especificadas
df = df.drop(columns=columnas_a_eliminar)
```

```python
df.head(10)
```

```python
#elimine las primeras filas del dataframe con el objetivo de limpiar la informacion que estaba en NAN en el principio
#y informacion que se repite que ya tenia claro como era.
df.drop(0, inplace=True)
df.drop(1, inplace=True)
df.drop(3, inplace=True)
df.drop(4, inplace=True)
df.drop(5, inplace=True)
df.drop(6, inplace=True)
df.drop(7, inplace=True)
```

```python
nuevos_nombres = ['ID', 'Lugar', 'Fecha', 'Temperatura (F°)', 'PM1', 'PM10', 'PM2.5']
# Cambiar los nombres de las columnas
df.columns = nuevos_nombres
```

```python
df
```

```python
# elimine otra de las filas que no aportaban informacion relevante dado que con el cambio de nombre de las columnas ya
# era algo repetitivo
df.drop(2, inplace=True)
```

```python
# Contar los valores nulos por columna
valores_nulos_por_columna = df.isnull().sum()

# Mostrar el resultado
print(valores_nulos_por_columna)
```

```python
 # en estos pasos procedi a llenar los espacios vacios del data frame con la moda de cada columna
 # por ejemplo en PM1 el dato que mas se repite es el 6, entonces los 271 datos faltantes les puse ese valor
 # realize esto para todo las columnas restantes
 # Calcular la moda de la columna 'PM1'
moda = df['PM1'].mode()[0]
print(moda)
# Llenar datos con la moda
df['PM1'].fillna(moda, inplace=True)

```

```python
# Calcular la moda de la columna 'PM10'
moda = df['PM10'].mode()[0]
# Llenar datos con la moda
df['PM10'].fillna(moda, inplace=True)
```

```python
# Calcular la moda de la columna 'PM2.5'
moda = df['PM2.5'].mode()[0]
# Llenar datos con la moda
df['PM2.5'].fillna(moda, inplace=True)
```

```python
# Calcular la moda de la columna 'temperatura'
moda = df['Temperatura (F°)'].mode()[0]
# Llenar datos con la moda
df['Temperatura (F°)'].fillna(moda, inplace=True)
```

```python
# Contar los valores nulos por columna
valores_nulos_por_columna = df.isnull().sum()

# Mostrar el resultado
print(valores_nulos_por_columna)
```

```python
import matplotlib.pyplot as plt
```

```python
# los datos de la columna fecha los procedi a pasar el tipo datatime con el objetivo de que los graficos fueran mas
#faciles de interpretar
df['Fecha'] = pd.to_datetime(df['Fecha']).astype(int)
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
# pase los datos a tipo numero para poder realizar las graficas pensadas
df['PM1'] = pd.to_numeric(df['PM1'], errors='coerce')
df['PM10'] = pd.to_numeric(df['PM10'], errors='coerce')
df['PM2.5'] = pd.to_numeric(df['PM2.5'], errors='coerce')
```

```python
df.head(10)
```

```python
# ajuste la columna fecha y los datos datatime para que se vieran de manera mas clara y entendibles
# anterior mente me daba un numero en nanosegundos que no eran muy claros
import datetime
df['Fecha'] = pd.to_numeric(df['Fecha'], errors='coerce')  # Convertir a números
df['Fecha'] = pd.to_datetime(df['Fecha'], unit='ns')  # Convertir a objetos datetime
df.head(10)
```

```python
# como se puede ver en el grafico de lineas como es de esperarse los datos con rangos mas grandes son los del PM10, }
#que es la particula mas grande de las 3, la que le sigue es la de PM2.5, la de PM1 nose aprecia demaciado dato que los datos
# de las demas caracteristicas lo opacan. en el grafico se puede observar que hay algunos datos atipicos en los meses
# de junio, julio, agosto y septiembre donde los niveles de Material particulado suben por encima de los limites en el que
# estas particulas empiezan a hacer daño para la salud de las personas. estos son datos de minitus exactos en donde se subieron
# de esta manera por lo que no signfica que el estado de la calidad de aire haya enpeorado por lo que se puede ver que bajan.

# Gráfico de líneas para PM1, PM10 y PM2.5
df.plot(x='Fecha', y=['PM1', 'PM10', 'PM2.5'], figsize=(12, 6))
plt.title('Calidad del aire')
plt.xlabel('Fecha')
plt.ylabel('Concentración')
plt.legend(['PM1', 'PM10', 'PM2.5'])
plt.grid(True)
plt.show()
```

```python
# en las siguientes graficas lo que se realizo fue separarlas por PM1, PM10 y PM2.5 para ver las tendencias
# en cada una de las caracteristicas. lo que se puede observar es que la mayoria de los datos estan en los rangos donde
# no presentan un impacto negativo para la salud de los seres vivos sin enmbargo en los 3 graficos se puede observar que
# los datos van subiendo en los ultimos meses se puede ver como hay mas acumulacion de datos por encima de 20, 50 y 20 respectivamente.
# indicandos que la calidad del aire se ha deteriorado un poco en los ultimos datos que sean tomado

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))

# Graficar PM1
axes[0].plot(df['Fecha'], df['PM1'], marker='o', linestyle='-')
axes[0].set_title('Concentración de PM1')
axes[0].set_xlabel('Fecha')
axes[0].set_ylabel('PM1')
axes[0].grid(True)

# Graficar PM10
axes[1].plot(df['Fecha'], df['PM10'], marker='o', linestyle='-')
axes[1].set_title('Concentración de PM10')
axes[1].set_xlabel('Fecha')
axes[1].set_ylabel('PM10')
axes[1].grid(True)

# Graficar PM2.5
axes[2].plot(df['Fecha'], df['PM2.5'], marker='o', linestyle='-')
axes[2].set_title('Concentración de PM2.5')
axes[2].set_xlabel('Fecha')
axes[2].set_ylabel('PM2.5')
axes[2].grid(True)

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
 # este grafico de dispersion entre PM1 y PM2.5 nos deja ver que ambas caracterisitcas estan muy relacionadas
 # entre mas sube PM1 el PM2.5 tambien sube. Por lo que un aumento en las particulas PM1 nos indica un aumento de igual
 # proporcion en las particulas PM2.5

 # Gráfico de dispersión para PM1 vs. PM2.5
plt.figure(figsize=(8, 6))
plt.scatter(df['PM1'], df['PM2.5'], c='b', label='PM1 vs. PM2.5')
plt.title('Relación entre PM1 y PM2.5')
plt.xlabel('PM1')
plt.ylabel('PM2.5')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# las particulas PM1 y PM10 tambien son directamente proporcionales.

# Gráfico de dispersión para PM1 vs. PM10
plt.figure(figsize=(8, 6))
plt.scatter(df['PM1'], df['PM10'], c='b', label='PM1 vs. PM10')
plt.title('Relación entre PM1 y PM10')
plt.xlabel('PM1')
plt.ylabel('PM10')
plt.legend()
plt.grid(True)
plt.show()
```
